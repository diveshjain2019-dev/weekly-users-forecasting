```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Srima (3).csv')
```

```python
df = pd.DataFrame(data)
df['CREATED_DT'] = pd.to_datetime(df['CREATED_DT'])
df.set_index('CREATED_DT', inplace=True)
```

```python
data= data[:-2]
```

```python
data.tail()
```

```python
data['CREATED_DT'] = pd.to_datetime(data['CREATED_DT'])

# Set the date column as the index
data.set_index('CREATED_DT', inplace=True)

# Log transformation to stabilize variance
data['OLD_TRADER_log'] = np.log1p(data['OLD_TRADER'])
```

```python
sarima_model = SARIMAX(data['OLD_TRADER_log'], order=(1, 1, 1), seasonal_order=(1, 1, 0, 52))
sarima_results = sarima_model.fit(disp=False)
```

```python
# # Forecast OLD_TRADER using SARIMA
# df['OLD_TRADER_log'] = np.log1p(df['OLD_TRADER'])  # Log-transform to handle non-stationarity
# sarima_model = SARIMAX(df['OLD_TRADER_log'], order=(1, 1, 1), seasonal_order=(1, 1, 0, 12))
# sarima_results = sarima_model.fit(disp=False)
```

```python
forecast_weeks = 15  # Forecasting for the next 12 weeks
forecast_index = pd.date_range(start=data.index[-1] + pd.Timedelta(weeks=1), periods=forecast_weeks, freq='W')
```

```python

forecast_old_trader_log = sarima_results.get_forecast(steps=forecast_weeks)
forecast_old_trader = np.expm1(forecast_old_trader_log.predicted_mean)  # Inverse log-transform
```

```python
seasonality_multipliers = {3: 3.0, 4: 3.25, 5: 4.75}
```

```python
adjusted_old_trader = []
for date, value in zip(forecast_index, forecast_old_trader):
    if (date.month == 3 and date.day >= 22) or (date.month == 4) or (date.month == 5 and date.day <= 25):
        multiplier = seasonality_multipliers.get(date.month, 1.75)  # Apply defined multipliers
    else:
        multiplier = 1.75  # Default multiplier for other dates
    adjusted_old_trader.append(value * multiplier)
```

```python
# Convert forecast to DataFrame
forecast_df = pd.DataFrame({'OLD_TRADER_forecast': adjusted_old_trader}, index=forecast_index)
```

```python
forecast_df = pd.DataFrame({
    'week': forecast_index,
    'Forecasted_Old_User_WAU': np.round(adjusted_old_trader)
})
forecast_df['Total_WAU'] = forecast_df['Forecasted_Old_User_WAU']
```

```python
# Combine with historical data
df.reset_index(inplace=True)
df.rename(columns={'CREATED_DT': 'Month'}, inplace=True)
historical_df = df[['Month', 'OLD_TRADER', 'NEW_TRADER']]
historical_df['Total_MAU'] = historical_df['OLD_TRADER']
```

```python
# Merge historical and forecasted data
final_df = pd.concat([
    historical_df[['Month', 'OLD_TRADER', 'NEW_TRADER', 'Total_MAU']],
    forecast_df
]).reset_index(drop=True)
```

```python
final_df.to_csv('FINAL_DF_WEEKLY.csv')
```

```python
# Plot Total MAU
plt.figure(figsize=(12, 6))
plt.plot(final_df['Month'], final_df['Total_MAU'], marker='o', label='WEEK')
plt.axvline(x=pd.Timestamp('2024-12-01'), color='red', linestyle='--', label='Forecast Start')
plt.title('WEEK (Historical and Forecasted)')
plt.xlabel('Month')
plt.ylabel('WEEK')
plt.legend()
plt.grid()
plt.savefig('WEEK (Historical and Forecasted).png')
plt.show()
```
